Question 1
(1). \begin{align*}
        E[U(x)] = E[x - \frac{\alpha x^2}{2}] = \mu - \frac{\alpha}{2}(\mu^2 + \sigma^2) = \mu - \frac{\alpha}{2}\mu^2 - \frac{\alpha}{2}\sigma^2$.
     \end{align*}
(2). \begin{align*}
        x_{CE} &= U^{-1}(\mu - \frac{\alpha}{2}\mu^2 - \frac{\alpha}{2}\sigma^2)\\
        &= \alpha^{-1}\cdot\big(1 - \big(1 - 2\alpha(\mu - \frac{\alpha}{2}\mu^2 - \frac{\alpha}{2}\sigma^2)\big)^{\frac{1}{2}} \big).
     \end{align*}
(3). \begin{align*}
        \pi_{A} &= \mu - x_{CE}\\
        &= \mu - \alpha^{-1}\cdot\big(1 - \big(1 - 2\alpha(\mu - \frac{\alpha}{2}\mu^2 - \frac{\alpha}{2}\sigma^2)\big)^{\frac{1}{2}} \big).
     \end{align*}
(4). Suppose we invest $z$ (unit: 1 million dollar) initially. Then the money $x \sim N(1 + \mu z + r(1-z), z^2 \sigma^2)$ and thus
     \begin{align*}
        x_{CE} &= \alpha^{-1}\cdot\big(1 - \big(1 - 2\alpha(1 + r + (\mu - r)z - \frac{\alpha}{2}(1 + r + (\mu - r)z)^2 - \frac{\alpha}{2}z^2\sigma^2)\big)^{\frac{1}{2}} \big)\\
        &= \alpha^{-1}\cdot\big(1 - \big(1 - 2\alpha((1+r)(1-\frac{\alpha}{2}(1+r)) + (1-a(1+r))(\mu - r)z - \frac{\alpha}{2}((\mu - r)^2 + \sigma^2)z^2)\big)^{\frac{1}{2}} \big).
     \end{align*}
And we have
     \begin{align*}
        z^* = \frac{(1 - \alpha(1+r))(\mu - z)}{\alpha((\mu - r)^2 + \sigma^2)} = \frac{\mu - r}{\alpha((\mu - r)^2 + \sigma^2)} - \frac{(1+r)(\mu - r)}{(\mu - r)^2 + \sigma^2}.
     \end{align*}

```python
from rl.gen_utils.plot_funcs import plot_list_of_curves

f = lambda al, mu, r, sig2: (mu - r)/(al*((mu - r)**2 + sig2)) - (1+r)*(mu - r)/((mu - r)**2 + sig2)
mu = [1.2, 2, 2]
r = [1, 1, 1]
sig2 = [1, 1, 1.5]

x_val = [s / 10 for s in range(1, 100)]

z_val = [[f(s, mu[i], r[i], sig2[i]) for s in x_val] for i in range(len(mu))]

plot_list_of_curves([x_val] * 3,
                    [z_val[0], z_val[1], z_val[2]],
                    ["r", "b", "g"],
                    [
                         r"$\alpha = 1.2, r = 1, \sigma^2 = 1$",
                         r"$\alpha = 2, r = 1, \sigma^2 = 1$",
                         r"$\alpha = 2, r = 1, \sigma^2 = 1.5$"
                     ],
                    r"$\alpha$",
                    r"$z$",
                    r"Plot: z - $\alpha$")
```

         E[\log(W)] = \log(W_0) + p\cdot \log(1 + \alpha f) + q\cdot \log(1 - \beta f).
     \end{align*}
(4). \begin{align*}
         \frac{d}{d f} E[\log(W)] = \frac{p \alpha}{1 + \alpha f} - \frac{q \beta}{1 - \beta f}.
